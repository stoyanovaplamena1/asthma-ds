**Overall Demographic Picture**

**Physical Activity**

**Diet Quality**

**Pollution, Pollen, and Dust Exposure**

**Overall Lifestyle Picture**

Next step is to explore the **medical history**.

```python
med_features = ['PetAllergy', 'FamilyHistoryAsthma', 'HistoryOfAllergies',
                'Eczema', 'HayFever', 'GastroesophagealReflux']

plt.figure(figsize=(12,8))
for i, col in enumerate(med_features, 1):
    plt.subplot(2,3,i)
    sns.countplot(x=col, data=asthma_df, palette="pastel", hue = col, legend = False)
    plt.title(f"{col} Distribution (0=No, 1=Yes)")
plt.tight_layout()
plt.show()
```

**Family History of Asthma**
Roughly 30% of patients have a family history of asthma. Thatâ€™s a significant portion, and it highlights the genetic or hereditary component of the disease.

**History of Allergies**
Around 40% of patients have a history of allergies. This is one of the larger groups, and it makes sense since asthma is often linked with allergic conditions.

**Eczema**
Nearly 19% of patients have eczema. This is important because eczema is part of the â€œatopic marchâ€ (eczema â†’ allergies â†’ asthma), which doctors often look for.

**Hay Fever**
About a quarter of patients (25%) have hay fever. Like eczema, itâ€™s another common allergic condition that can be connected with asthma.

**Gastroesophageal Reflux (GERD)**
Around 16% of patients have reflux issues. GERD is not an allergy, but it is known to worsen asthma symptoms in some cases.

**Overall Medical History Picture**
The dataset shows a strong presence of allergy-related conditions, with many patients reporting allergies, hay fever, or eczema. Family history of asthma is also fairly common. Together, these factors provide a realistic clinical profile, since asthma often appears alongside other allergic conditions.

```python
# --- FEV1 ---
plt.figure(figsize=(6,4))
sns.histplot(asthma_df['LungFunctionFEV1'], bins=30, kde=True, color="skyblue")
plt.axvline(asthma_df['LungFunctionFEV1'].mean(), color='red', linestyle='--', label='Mean')
plt.title("FEV1 Distribution")
plt.xlabel("FEV1 (Liters)")
plt.ylabel("Count")
plt.legend()
plt.show()

# --- FVC ---
plt.figure(figsize=(6,4))
sns.histplot(asthma_df['LungFunctionFVC'], bins=30, kde=True, color="orange")
plt.axvline(asthma_df['LungFunctionFVC'].mean(), color='red', linestyle='--', label='Mean')
plt.title("FVC Distribution")
plt.xlabel("FVC (Liters)")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# --- Create new column ---
asthma_df['FEV1_FVC_ratio'] = asthma_df['LungFunctionFEV1'] / asthma_df['LungFunctionFVC']

# --- Quick stats ---
print(asthma_df['FEV1_FVC_ratio'].describe())

# --- Histogram with diagnostic threshold ---
plt.figure(figsize=(6,4))
sns.histplot(asthma_df['FEV1_FVC_ratio'], bins=30, kde=True, color="teal")
plt.axvline(asthma_df['FEV1_FVC_ratio'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(0.7, color='black', linestyle=':', label='Diagnostic cutoff (0.7)')
plt.title("Distribution of FEV1/FVC Ratio")
plt.xlabel("FEV1/FVC Ratio")
plt.ylabel("Count")
plt.legend()
plt.show()

# --- Boxplot ---
plt.figure(figsize=(4,5))
sns.boxplot(y=asthma_df['FEV1_FVC_ratio'], color="lightblue")
plt.axhline(0.7, color='black', linestyle=':', label='Diagnostic cutoff (0.7)')
plt.title("FEV1/FVC Ratio Boxplot")
plt.ylabel("FEV1/FVC Ratio")
plt.legend()
plt.show()
```

**FEV1 (Forced Expiratory Volume in 1 second)**
The average FEV1 is about 2.55 liters, with values ranging from around 1 to 4 liters. This spread looks realistic: lower values often reflect impaired lung function, while higher values suggest normal or strong capacity. 

**FVC (Forced Vital Capacity)**
The average FVC is about 3.74 liters, ranging from 1.5 to 6 liters. This is a wider spread than FEV1 and includes patients with both reduced and normal lung volumes. It looks consistent with expected physiological ranges.

**FEV1/FVC Ratio**
I calculated this ratio because it seems to be an important factor in diagnosing obstructive airway deseases. Clinically, a ratio below 0.7 is a key cutoff for making a diagnosis. The average ratio is 0.79, which is just above the clinical cutoff. The median is slightly lower at 0.68, meaning half of the patients fall below the threshold â€” a strong signal that many in this dataset show obstructive patterns.

The spread is wide: from about 0.17 (severely impaired lung function) up to 2.57 (which is unusually high and might indicate data artifacts or extreme cases). The 25th percentile (0.47) and 75th percentile (0.98) show that most patients lie between moderate obstruction and normal function.

**Overall Lung Function Picture**
Both FEV1 and FVC fall within realistic medical ranges and show good variation. The implied FEV1/FVC ratio being near the diagnostic threshold adds confidence that this dataset captures the spectrum from healthy to impaired lung function.
